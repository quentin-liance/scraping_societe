```python
import pandas as pd
import numpy as np
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs
from urllib.request import URLopener
```

```python
# Enterprise list
urlList = [
"https://www.societe.com/bilan/hugel-et-fils-sa-915722144",
"https://www.societe.com/bilan/maison-louis-jadot-403231137",
"https://www.societe.com/bilan/caves-bailly-lapierre-778681122",
"https://www.societe.com/bilan/union-des-vignerons-des-cotes-du-rhone-302558622",
"https://www.societe.com/bilan/sca-les-vignerons-du-val-d-orbieu-775812258",
"https://www.societe.com/bilan/chateau-de-santenay-412618001"
]
# Prepare empty Array 
globalDf = []
my_ent = []

# 1  define own years range 

startYear = 2004
endYear = 2020
# 2  Create function to get website data (bilan) 
def gen_soup(url):
    req = Request(url, headers={'User-Agent': 'Safari'}) # changer le navigateur éventuellement (ça ne change pas grand chose)
    webpage = urlopen(req).read()
    return bs(webpage,"html.parser") # renvoie le code HTML

# 3 send enterprise data to array
for i in urlList:
    for j in range(startYear, endYear):
        url = i+f"{j}12311.html"
        print(j)
        print(url)
        datas = []
        urlText = gen_soup(url)
        tables_rows = urlText.findAll('table')
        for table in tables_rows:
            df = pd.read_html(str(table))
            datas.append(df)
        # get enterprise name
        placeSlash = i.rfind('/')+1
        placeDash = i.rfind('-')
        enterprise  = i[placeSlash:placeDash]
        my_ent.append(enterprise)
        infoSend  = {
            'enterprise':  enterprise, 
            'year':j, 
            'data':datas
            }
        #Send to global Array container
        globalDf.append(infoSend)
    print(globalDf)
    
```

```python
firms_name = []
firms_name.append(my_ent[0])
r = 0
for i in my_ent: 
    if firms_name[r] != i:
        firms_name.append(i)
        r += 1

my_infos = ['AI', 'AC', 'CapP', 'D', 'CA', 'Prod', 'Profit']

my_columns = [
    np.repeat(np.array(firms_name), len(my_infos)),
    np.tile(np.array(my_infos), len(firms_name))
]

my_dates = range(startYear, endYear)
df = pd.DataFrame(np.random.rand(len(my_dates), len(my_columns[0])), index=my_dates, columns=my_columns)
df
```

```python
for i in range(len(globalDf)): 
    name = globalDf[i].get('enterprise')
    year = globalDf[i].get('year')
    asset = globalDf[i].get('data')[0]
    liabilitie = globalDf[i].get('data')[1]
    income = globalDf[i].get('data')[2]
    
    df[name, 'AI'][year] = asset[0].iloc[2,1]
    df[name, 'AC'][year] = asset[0].iloc[7,1]
    df[name, 'CapP'][year] = liabilitie[0].iloc[2,1]
    df[name, 'D'][year] = liabilitie[0].iloc[5,1]
    df[name, 'CA'][year] = income[0].iloc[2,1]
    df[name, 'Prod'][year] = income[0].iloc[4,1]
    df[name, 'Profit'][year] = float(income[0].iloc[11,1])
    
df  
```

```python
df.to_csv('wine_maker_financial.csv')
```
